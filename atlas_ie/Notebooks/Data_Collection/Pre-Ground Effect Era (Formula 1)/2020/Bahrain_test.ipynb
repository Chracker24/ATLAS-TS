## Importing Necessary Libraries

## Getting one race data for an example 

```python
session = fastf1.get_session(2020, "Bahrain Grand Prix", "R")
session.load()

laps =  session.laps
laps.head()
```

```python
lap_list = list(laps.columns)
for i in range (len(lap_list)):
    print(lap_list[i])
```

## Making a Proper Data file and saving it in the Data Folders

```python
cols = [
    "Driver",
    "LapTime",
    "Stint",
    "Sector1Time",
    "Sector2Time",
    "Sector3Time",
    "Compound",
    "Team",
    "Deleted",
]

Bahrain_2020 = laps[cols].copy()
Bahrain_2020
```

### Adding the Race and Season tags for future Reference

```python
Bahrain_2020 = Bahrain_2020.dropna(subset=["LapTime"]).copy()
Bahrain_2020["Season"] = 2020
Bahrain_2020["Race"] = "Bahrain"
Bahrain_2020
```

### We can see that the LapTime, Sector1Time, Sector2Time and Sector3Time are in Deltas and not in Seconds. This is what I will change them into in the next step for a Time-Series Step

```python
Bahrain_2020["LapTime"] = Bahrain_2020["LapTime"].dt.total_seconds().copy()
Bahrain_2020["Sector1Time"] = Bahrain_2020["Sector1Time"].dt.total_seconds().copy()
Bahrain_2020["Sector2Time"] = Bahrain_2020["Sector2Time"].dt.total_seconds().copy()
Bahrain_2020["Sector3Time"] = Bahrain_2020["Sector3Time"].dt.total_seconds().copy()

Bahrain_2020.head()
```

### Small Precaution of taking out Deleted LapTimes

```python
Bahrain_2020 = Bahrain_2020[Bahrain_2020["Deleted"]==False].copy()
Bahrain_2020
```
